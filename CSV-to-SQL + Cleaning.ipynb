```python
# removing redundant variables and datapoints with no combatative use (languages)
df = df[['name', 'size', 'type', 'alignment', 'armor_class', 'hit_points', 'hit_dice', 'speed',
       'strength', 'dexterity', 'constitution', 'intelligence', 'wisdom',
       'charisma', 'proficiencies', 'damage_vulnerabilities',
       'damage_resistances', 'damage_immunities', 'condition_immunities',
       'senses', 'challenge_rating', 'special_abilities',
       'actions', 'legendary_actions', 'reactions', 'other_speeds']]
df.replace('[]', np.nan, inplace=True)
df.head()
```

```python
data = pd.DataFrame(df["speed"].value_counts())
# pd.set_option('display.max_rows', None)   // For showing all data forms
data
```

```python
df["speed"] = df["speed"].str.replace("{", '')
df["speed"] = df["speed"].str.replace("}", '')
df["speed"] = df["speed"].str.replace("'", '')
df["speed"] = df["speed"].str.replace(",", ' |')
df["speed"] = df["speed"].str.replace("walk", 'Walk')
df["speed"] = df["speed"].str.replace("fly", 'Fly')
df["speed"] = df["speed"].str.replace("swim", 'Swim')
df["speed"] = df["speed"].str.replace("climb", 'Climb')
df["speed"] = df["speed"].str.replace("hover", 'Hover')
df["speed"] = df["speed"].str.replace("burrow", 'Burrow')
data = pd.DataFrame(df["speed"].unique())
data
```

```python
data = pd.DataFrame(df["proficiencies"].value_counts())
# pd.set_option('display.max_rows', None)   // For showing all data forms
data
```

```python
df["proficiencies"] = df["proficiencies"].str.replace("[", '')
df["proficiencies"] = df["proficiencies"].str.replace("]", '')
df["proficiencies"] = df["proficiencies"].str.replace("{", '')
df["proficiencies"] = df["proficiencies"].str.replace("}", '')
df["proficiencies"] = df["proficiencies"].str.replace("',", ' ')
df["proficiencies"] = df["proficiencies"].str.replace("'", ' ')
df["proficiencies"] = df["proficiencies"].str.replace(",", ' |')
df["proficiencies"] = df["proficiencies"].str.title()
df["proficiencies"] = df["proficiencies"].str.replace("Url :", '')
df["proficiencies"] = df["proficiencies"].str.replace("Name : ", '')
df["proficiencies"] = df["proficiencies"].str.replace("Value : ", '+')
df["proficiencies"] = df["proficiencies"].str.replace("/Api/Proficiencies/", '')
data = pd.DataFrame(df["proficiencies"].value_counts())
data
```

```python
data = pd.DataFrame(df["other_speeds"].value_counts())
# pd.set_option('display.max_rows', None)
data
```

```python
df["other_speeds"] = df["other_speeds"].str.replace("{", '')
df["other_speeds"] = df["other_speeds"].str.replace("}},", ' |')
df["other_speeds"] = df["other_speeds"].str.replace("[", '')
df["other_speeds"] = df["other_speeds"].str.replace("]", '')
df["other_speeds"] = df["other_speeds"].str.replace("}", '')
df["other_speeds"] = df["other_speeds"].str.replace("'", '')
df["other_speeds"] = df["other_speeds"].str.replace(",", '')
df["other_speeds"] = df["other_speeds"].str.replace("walk", 'Walk')
df["other_speeds"] = df["other_speeds"].str.replace("climb", 'Climb')
df["other_speeds"] = df["other_speeds"].str.replace("form: hyrid", 'Hybrid')
df["other_speeds"] = df["other_speeds"].str.replace("form: bear", 'Bear')
df["other_speeds"] = df["other_speeds"].str.replace("form: boar", 'Boar')
df["other_speeds"] = df["other_speeds"].str.replace("form: tiger", 'Tiger')
df["other_speeds"] = df["other_speeds"].str.replace("form: wolf", 'Wolf')
df["other_speeds"] = df["other_speeds"].str.replace("speed", 'Speeds')
data = pd.DataFrame(df["other_speeds"].value_counts())
data
```
