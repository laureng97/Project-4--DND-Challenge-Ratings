```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./Source Files/monsters.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
# removing redundant variables and datapoints with no combatative use (languages)
df = df[['name', 'size', 'type', 'alignment', 'armor_class', 'hit_points', 'hit_dice', 'speed',
       'strength', 'dexterity', 'constitution', 'intelligence', 'wisdom',
       'charisma', 'proficiencies', 'damage_vulnerabilities',
       'damage_resistances', 'damage_immunities', 'condition_immunities',
       'senses', 'challenge_rating', 'special_abilities',
       'actions', 'legendary_actions', 'reactions', 'other_speeds']]
df.replace('[]', np.nan, inplace=True)
df.head()
```

# Easy API Stripping
Can be easily processed using get_dummies
- damage_vulnerabilities 
- damage_resistances 
- damage_immunities 
- condition_immunities 

```python
data = pd.DataFrame(df["damage_vulnerabilities"].value_counts())
# pd.set_option('display.max_rows', None)   // For showing all data forms
data
```

```python
df["damage_vulnerabilities"] = df["damage_vulnerabilities"].str.replace("[", '')
df["damage_vulnerabilities"] = df["damage_vulnerabilities"].str.replace("]", '')
df["damage_vulnerabilities"] = df["damage_vulnerabilities"].str.replace("'", '')
df["damage_vulnerabilities"] = df["damage_vulnerabilities"].str.replace(",", ' |')
df["damage_vulnerabilities"] = df["damage_vulnerabilities"].str.title()
data = pd.DataFrame(df["damage_vulnerabilities"].value_counts())
data
```

```python
data = pd.DataFrame(df["damage_resistances"].value_counts())
# pd.set_option('display.max_rows', None)   // For showing all data forms
data
```

```python
df["damage_resistances"] = df["damage_resistances"].str.replace("[", '')
df["damage_resistances"] = df["damage_resistances"].str.replace("]", '')
df["damage_resistances"] = df["damage_resistances"].str.replace("', ", ' | ')
df["damage_resistances"] = df["damage_resistances"].str.replace("'", '')
df["damage_resistances"] = df["damage_resistances"].str.replace('"', '')
df["damage_resistances"] = df["damage_resistances"].str.title()
data = pd.DataFrame(df["damage_resistances"].value_counts())
data
```

```python
data = pd.DataFrame(df["damage_immunities"].value_counts())
# pd.set_option('display.max_rows', None)   // For showing all data forms
data
```

```python
df["damage_immunities"] = df["damage_immunities"].str.replace("[", '')
df["damage_immunities"] = df["damage_immunities"].str.replace("]", '')
df["damage_immunities"] = df["damage_immunities"].str.replace("', ", ' | ')
df["damage_immunities"] = df["damage_immunities"].str.replace("'", '')
df["damage_immunities"] = df["damage_immunities"].str.replace('"', '')
df["damage_immunities"] = df["damage_immunities"].str.title()
data = pd.DataFrame(df["damage_immunities"].value_counts())
data
```

```python
data = pd.DataFrame(df["condition_immunities"].value_counts())
# pd.set_option('display.max_rows', None)   // For showing all data forms
data
```

```python
df["condition_immunities"] = df["condition_immunities"].str.replace("[", '')
df["condition_immunities"] = df["condition_immunities"].str.replace("]", '')
df["condition_immunities"] = df["condition_immunities"].str.replace("}, ", ' | ')
df["condition_immunities"] = df["condition_immunities"].str.replace("{", '')
df["condition_immunities"] = df["condition_immunities"].str.replace("}", '')
df["condition_immunities"] = df["condition_immunities"].str.replace("'", '')
df["condition_immunities"] = df["condition_immunities"].str.replace("name: ", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(", url: /api/conditions/", ' ')
df["condition_immunities"] = df["condition_immunities"].str.replace(" poisoned", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" charmed", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" frightened", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" grappled", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" paralyzed", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" petrified", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" prone", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" restrained", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" stunned", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" exhaustion", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" unconscious", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" blinded", '')
df["condition_immunities"] = df["condition_immunities"].str.replace(" deafened", '')
data = pd.DataFrame(df["condition_immunities"].value_counts())
data
```

```python
data = pd.DataFrame(df["proficiencies"].value_counts())
# pd.set_option('display.max_rows', None)   // For showing all data forms
data
```

```python
df["proficiencies"] = df["proficiencies"].str.replace("[", '')
df["proficiencies"] = df["proficiencies"].str.replace("]", '')
df["proficiencies"] = df["proficiencies"].str.replace("{", '')
df["proficiencies"] = df["proficiencies"].str.replace("}", '')
df["proficiencies"] = df["proficiencies"].str.replace("',", ' ')
df["proficiencies"] = df["proficiencies"].str.replace("'", ' ')
df["proficiencies"] = df["proficiencies"].str.replace(",", ' |')
df["proficiencies"] = df["proficiencies"].str.title()
df["proficiencies"] = df["proficiencies"].str.replace("Url :", '')
df["proficiencies"] = df["proficiencies"].str.replace("Name : ", '')
df["proficiencies"] = df["proficiencies"].str.replace("Value : ", '+')
df["proficiencies"] = df["proficiencies"].str.replace("/Api/Proficiencies/", '')
data = pd.DataFrame(df["proficiencies"].value_counts())
data
```
